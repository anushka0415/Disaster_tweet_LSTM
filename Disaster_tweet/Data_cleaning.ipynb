```python
import re, string, unicodedata
```

```python
def hashtags(texts):
    hash_tags = re.findall('#([a-zA-z]+)', texts) or re.findall(r'<HASHTAG>\s*([a-zA-Z]+)', texts)
    lower = [tags.lower() for tags in hash_tags]
    return lower
df['hashtags'] = df['tweets'].apply(hashtags)
```

```python
df['hashtags'] = df['hashtags'].apply(lambda x: x if x else 'None')
```

```python
df
```

```python
def remove_placeholders(text):
    cleaned_text = re.sub(r'<USER>|<URL>|<HASHTAG>', '', text) or re.sub(r'@[\w_]+', '', text)
    return cleaned_text
df['tweets'] = df['tweets'].apply(remove_placeholders)
```

```python
def remove_urls(text):
    url_pattern = r'https?://\S+|www\.\S+'
    cleaned_text = re.sub(url_pattern, '', text)
    return cleaned_text

df['tweets'] = df['tweets'].apply(remove_urls)
```

```python
def clean(text):
    text = re.sub('[^A-Za-z]+',' ',text)
    return text

df['tweets'] = df['tweets'].apply(clean)
df['tweets'] = df['tweets'].apply(lambda x:x.lower())
```

```python
df.head()
```

```python
def remove_rt(text):
    text = text.replace('rt', '').strip()
    return text
df['tweets'] = df['tweets'].apply(remove_rt)
```

```python
file_path = 'cleaned.csv'
df.to_csv(file_path, index=False)
```
