```python
import pandas as pd
import matplotlib.pyplot as plt



def plot_character_length_histogram(text):
    text_length = len(text)
    return text_length

# Apply the function to the 'tweets' column and create a new column with character lengths
df['tweet_length'] = df['tweets'].apply(plot_character_length_histogram)

# Plot a histogram of the tweet lengths
plt.hist(df['tweet_length'], bins=30) 
plt.xlabel('Character Length')
plt.ylabel('Frequency')
plt.title('Character Length Histogram')
plt.show()



```

```python
import pandas as pd
import matplotlib.pyplot as plt


df['tweets'] = df['tweets'].astype(str)

def plot_word_number_histogram(text):
    word_lengths = text.split()  
    num_words = len(word_lengths) 
    return num_words


df['word_count'] = df['tweets'].apply(plot_word_number_histogram)


plt.hist(df['word_count'], bins=30) 
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.title('Word Count Histogram')
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np
from nltk.corpus import stopwords

def plot_top_stopwords_barchart(text):
    stop=set(stopwords.words('english'))
    
    new= text.str.split()
    new=new.values.tolist()
    corpus=[word for i in new for word in i]
    from collections import defaultdict
    dic=defaultdict(int)
    for word in corpus:
        if word in stop:
            dic[word]+=1
            
    top=sorted(dic.items(), key=lambda x:x[1],reverse=True)[:10] 
    x,y=zip(*top)
    plt.bar(x,y)
    
plot_top_stopwords_barchart(df.tweets)
```

```python
import gensim
from nltk.stem import WordNetLemmatizer,PorterStemmer
from nltk.tokenize import word_tokenize
import pyLDAvis.gensim


def get_lda_objects(text):
    nltk.download('stopwords')    
    stop=set(stopwords.words('english'))

    
    def _preprocess_text(text):
        corpus=[]
        stem=PorterStemmer()
        lem=WordNetLemmatizer()
        for news in text:
            words=[w for w in word_tokenize(news) if (w not in stop)]

            words=[lem.lemmatize(w) for w in words if len(w)>2]

            corpus.append(words)
        return corpus
    
    corpus=_preprocess_text(text)
    
    dic=gensim.corpora.Dictionary(corpus)
    bow_corpus = [dic.doc2bow(doc) for doc in corpus]
    
    lda_model =  gensim.models.LdaMulticore(bow_corpus, 
                                   num_topics = 4, 
                                   id2word = dic,                                    
                                   passes = 10,
                                   workers = 2)
    
    return lda_model, bow_corpus, dic

def plot_lda_vis(lda_model, bow_corpus, dic):
    pyLDAvis.enable_notebook()
    vis = pyLDAvis.gensim.prepare(lda_model, bow_corpus, dic)
    return vis
```

```python
lda_model, bow_corpus, dic = get_lda_objects(df.tweets)
lda_model.show_topics()
```

```python
from textblob import TextBlob
def polarity(text):
    return TextBlob(text).sentiment.polarity

df['polarity_score']=df['tweets'].apply(lambda x : polarity(x))
df['polarity_score'].hist()
```
